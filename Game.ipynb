```python
import json
import random

class Character():  #Parent Class
    def __init__ (self, Name):
        self.Name = Name 
        self.HP = 100 
        self.AD = 10
        self.Shield = 0
        self.bag = []

    def setLevel_one(self): #Resets stats if you want
        self.HP = 100 
        self.XP = 0
        self.AD = 7
        self.Shield = 50

    def get_stats(self):    #Prints the character Stats  
        print("Name: ", self.Name)
        print("HP: ",self.HP)
        print("AD: ", self.AD)
        print("Shield: ", self.Shield)

    def get_bag(self):
        print(f"Items in {self.Name}'s bag: ")
        print(json.dumps(self.bag,indent = 3))

    def add_item(self, item):
        for itm in item:
            self.bag.append(itm)

    def discard_item(self, delete):
        print(self.bag)
        for item in self.bag:
            if item["name"] == delete:
                self.bag.remove(item)
    
    def item_buffs(self):
        for item in self.bag:
            self.HP = self.HP + item["heal"]
            self.Shield = self.Shield + item["shield"]
            self.AD = self.AD + item["AD"]

#######################################################

class Attacker(Character): #Child class from Character class
    def __init__(self, Name,):
        super().__init__(Name) 
        self.Name = Name 
        self.AD = 12
        self.HP = 90 
        self.Shield = 0

######################################################

class Support(Character): #Child class from Character class
    def __init__(self, Name):
        super().__init__(Name) 
        self.Name = Name 
        self.AD = 7
        self.HP = 110 
        self.Shield = 50 

#######################################################

class Tank(Character): #Child class from Character class
    def __init__(self, Name):
        super().__init__(Name) 
        self.Name = Name 
        self.AD = 8
        self.HP = 110 
        self.Shield = 55 

#######################################################

class Enemy(Character): #Child class from Character class
    def __init__(self, Name):
        super().__init__(Name) 
        self.Name = Name 
        self.AD = 8
        self.HP = 200 
        self.Shield = 55 

######################################################
item = [
    
    {'name': 'ball','AD':5,'heal':0,'shield':0},
    {'name': 'stapler','AD':7,'heal':0,'shield':0},
    {'name': 'gum','AD':0,'heal':0,'shield':10},
    {'name': 'spoon','AD':50,'heal':50,'shield':50},
    {'name': 'pencil','AD':6,'heal':0,'shield':0},
    {'name': 'cilantro','AD':0,'heal':20,'shield':0},
    {'name': 'tomato with tajin','AD':0,'heal':10,'shield':10},
    {'name': 'knife','AD':10,'heal':0,'shield':0}
    
    ]

####################################################

def game():
    chance = 0
    cont = "y"
    gameplay = "y"
    while (cont == "y"):
        game = input("Hello, ready to play? (y/n)?")
        
        if game == "y":
            name = input("Please choose a character name ")
            type = input("Please choose a type of character: attacker, tank, or support ") 
            if type == "attacker":
                player = Attacker(name)
            elif type == "tank":
                player = Tank(name)
            elif type == "support":
                player = Support(name)
            
            while gameplay == "y":
                options = input("Welcome, what would you like to do?\nExplore:\nInspect bag:\nQuit:") 
                    
                if options.lower() == "explore":
                    chance = random.randint(1,3)
                    if chance == 1:
                        print("Boss encounter")
                        boss = Enemy("Coffee Bean & tea leaff")
                        player.item_buffs()
                        boss_fight(player,boss)
                    elif chance == 2:
                        print("Item found!")
                        player.add_item(random.sample(item,1))
                    elif chance == 3:
                            print("Keep it movin, nothing to see here! ")
    
                elif options.lower() == "inspect bag":
                    bag_action = input("View Bag or Discard item? (discard/view) ")
                    if bag_action.lower() == "discard":
                        trash = input("Select an item to discard") 
                        player.discard_item(trash)
                    elif bag_action == "view":
                        player.get_bag()
            
                elif options.lower() == "quit":
                    player_one = {      #Creating a dictionary with the stats of the current player and dumping them into a json
                        'Name':player.Name,
                        'HP':player.HP,
                        'AD':player.AD,
                        'Shield':player.Shield,
                        'Bag':player.bag
                                }
                    with open("Game1.json","w") as f:
                        json.dump(player_one,f)
                        game = "n"
                        cont ="n"
                        
                    break

        elif game =="n" :
            cont == "n"
            break
            
#######################################################
        
def boss_fight(player,boss):

    battle = "y"

    while battle == "y": 
        print("Your current stats are: \nHP:",player.HP, "\nAD:", player.AD,"\nShield:",player.Shield)
        print("The boss stats are: \nHP:",boss.HP, "\nAD:", boss.AD,"\nShield:",boss.Shield)
        choice = input("Ready to fight? (y/n) ")
        turns = 0

        if choice == "y":
            
            startHP = player.HP
            bossHP = boss.HP

            player.HP = player.HP + player.Shield
            boss.HP = boss.HP + boss.Shield
            print(player.HP , boss.HP)

            while True:
                turns = turns+1
                if player.HP > 0:
                    if boss.HP > 0:
                        player.HP = player.HP - boss.AD
                        boss.HP = boss.HP - player.AD 
                        if boss.HP <= 0 :
                            print("You Win")
                            print("Number of turns: ",turns)
                            battle = "n"
                            boss.HP = bossHP
                            break
                        elif player.HP <= 0:
                            print("You lose")
                            print("Number of turns: ",turns)
                            battle = "n"
                            player.HP = startHP
                            break
                    #else: break
                #else: break
        elif choice == "n": 
            battle = "n"




game()
```
